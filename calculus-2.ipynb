```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = sp.exp(-0.3 * x) * sp.sin(3 * x)

integral = sp.integrate(f, (x, 0, 6 * sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(0, 6 * np.pi, 200)
y_vals = np.exp(-0.3 * x_interval) * np.sin(3 * x_interval)

plt.figure(figsize=(10, 6))
plt.plot(x_interval, y_vals, color="blue")

intervals = [
    (0, np.pi / 3), 						# 0deg   -> 60deg
    (2 * np.pi / 3, np.pi), 				# 120deg -> 180deg
    (4 * np.pi / 3, np.pi + 2 * np.pi / 3), # 240deg -> 300deg
    (2 * np.pi, 2 * np.pi + np.pi / 3), 	# 360deg -> 420deg
    (8 * np.pi / 3, 3 * np.pi), 			# 480deg -> 540deg
    (10 * np.pi / 3, 11 * np.pi / 3),		# 600deg -> 660deg
    (4 * np.pi, 4 * np.pi + (np.pi / 3)),	# 720deg -> 780deg
    (14 * np.pi / 3, 5 * np.pi),			# 840deg -> 900deg
    (16 * np.pi / 3, 17 * np.pi / 3), 		# 960deg -> 1020deg
]

for start, end in intervals:
    x_fill = np.linspace(start, end, 200)
    y_fill = np.exp(-0.3 * x_fill) * np.sin(3 * x_fill)
    plt.fill_between(x_fill, y_fill, alpha=0.3)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = sp.sin(x) + sp.sin(3 * x) + (1 / 2) * sp.sin(5 * x)

integral = sp.integrate(f, (x, 0, 2 * sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(0, 2 * np.pi, 200)
y_vals = np.sin(x_interval) + np.sin(3 * x_interval) + (1 / 2) * np.sin(5 * x_interval)

plt.figure(figsize=(10, 6))
plt.plot(x_interval, y_vals, color="blue")

xy_fill = np.linspace(0, np.pi, 200)
plt.fill_between(
    xy_fill, np.sin(xy_fill) + np.sin(3 * xy_fill) + (1 / 2) * np.sin(5 * xy_fill),
    alpha=0.3
)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = (x**2) * sp.sin(x)

integral = sp.integrate(f, (x, 0, sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(0, np.pi, 200)
y_vals = (x_interval**2) * np.sin(x_interval)

plt.figure(figsize=(10, 6))
plt.plot(x_interval, y_vals, color="blue")

xy_fill = np.linspace(0, np.pi, 200)
plt.fill_between(
    xy_fill, (xy_fill**2) * np.sin(xy_fill),
    alpha=0.3
)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = (1 + 0.5 * sp.sin(x)) * sp.cos(4 * x)

integral = sp.integrate(f, (x, 0, 4 * sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(0, 4 * np.pi, 200)
y_vals = (1 + 0.5 * np.sin(x_interval)) * np.cos(4 * x_interval)

plt.figure(figsize=(10, 6))
plt.plot(x_interval, y_vals, color="blue")

xy_fill = np.linspace(0, 4 * np.pi, 200)
plt.fill_between(xy_fill, (1 + 0.5 * np.sin(xy_fill)) * np.cos(4 * xy_fill), alpha=0.3)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

$$
f(x) = 
\begin{cases}
	\sin(x) & \text{if } 0 \le x < \pi \\
	0.5\cos(x) & \text{if } \pi \le x \le 2\pi
\end{cases}
\\
A = \int_{0}^{\pi} |\sin(x)| \space dx + \int_{\pi}^{2\pi} |0.5\sin(2x)| \space dx
$$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols("x")
f1 = sp.sin(x)
f2 = 0.5 * sp.cos(x)

integral1 = sp.integrate(f1, (x, 0, 2 * sp.pi))
integral2 = sp.integrate(f2, (x, 0, 2 * sp.pi))

print(f"Hasil integral 1: {integral1}")
print(f"Hasil desimal 1: {float(integral1)}")

print(f"Hasil integral 2: {integral2}")
print(f"Hasil desimal 2: {float(integral2)}")

x_interval_sin = np.linspace(0, 2 * np.pi, 200)
x_interval_cos = np.linspace(0, 2 * np.pi, 200)

y_vals_sin = np.sin(x_interval_sin)
y_vals_cos = 0.5 * np.cos(x_interval_cos)

plt.figure(figsize=(10, 6))
plt.plot(x_interval_cos, y_vals_cos, color="red")
plt.plot(x_interval_sin, y_vals_sin, color="blue")

plt.fill_between(
    np.linspace(0, np.pi, 200),
    np.sin(np.linspace(0, np.pi, 200)),
    alpha=0.3
)
plt.fill_between(
    np.linspace(0, 2 * np.pi / 4, 200),
    0.5 * np.cos(np.linspace(0, 2 * np.pi / 4, 200)),
    alpha=0.3,
)
plt.fill_between(
    np.linspace(2 * np.pi - np.pi / 2, 2 * np.pi, 200),
    0.5 * np.cos(np.linspace(2 * np.pi - np.pi / 2, 2 * np.pi, 200)),
    alpha=0.3,
)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
