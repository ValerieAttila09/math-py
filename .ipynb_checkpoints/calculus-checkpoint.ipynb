```python
import math
import numpy as np

# Sudut dalam radian
theta = math.pi / 4  # 45 derajat

# Fungsi trigonometri dasar
print("sin(θ) =", math.sin(theta))
print("cos(θ) =", math.cos(theta))
print("tan(θ) =", math.tan(theta))

# Jika ingin pakai derajat langsung dengan numpy
angle_deg = 30
print("sin(30°) =", np.sin(np.deg2rad(angle_deg)))
```

```python
import sympy as sp

# Definisikan variabel
x = sp.symbols('x')

# Fungsi
f = sp.sin(x) * sp.cos(x)

# Turunan
f_prime = sp.diff(f, x)
print("Turunan f(x):", f_prime)

# Integral
f_integral = sp.integrate(f, x)
print("Integral f(x):", f_integral)

# Limit
limit_value = sp.limit(sp.sin(x)/x, x, 0)
print("Limit sin(x)/x saat x→0:", limit_value)
```

$$
f(x) = \sin(x) . \cos(x)
$$
$$
f'(x) = -\sin(x)^2 + \cos(x)^2
$$
$$
\int \frac{\sin(x)^2}{2}
$$
$$
\lim_{x\rightarrow0} \frac{\sin(x)}{x} = 1
$$

```python
import matplotlib.pyplot as plt

x_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
y_vals = np.sin(x_vals)

plt.plot(x_vals, y_vals)
plt.title("Grafik sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
