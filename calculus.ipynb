```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# ==================================== #
# 1. Definisikan variabel simbolik     #
# ==================================== #
x = sp.symbols("x")
f = x**2

# ==================================== #
# 2. Hitung integral tentu             #
# ==================================== #
integral = sp.integrate(f, (x, 0, 2))

print("Hasil integral", integral)
print("Hasil desimal", float(integral))

# ==================================== #
# 3. Buat data grafik                  #
# ==================================== #
x_vals = np.linspace(0, 2.5, 200)
y_vals = x_vals**2

# ==================================== #
# 4. Plot grafik                       #
# ==================================== #
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2", color="blue")

# ==================================== #
# 5. Shade area integrasi              #
# ==================================== #
x_fill = np.linspace(0, 2, 200)
y_fill = x_fill**2

plt.fill_between(x_fill, y_fill, alpha=0.3)

# garis batang integral
plt.axvline(0)
plt.axvline(2)

plt.title("Grafik Trigonometri")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = 2 * sp.sin(x + (sp.pi * 0.25))

integral = sp.integrate(f, (x, -1 * sp.pi, sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(-np.pi, np.pi, 200)
y_vals = 2 * np.sin(x_interval + (np.pi * 0.25))

plt.figure(figsize=(8, 6))
plt.plot(x_interval, y_vals, color="blue")

x_arg = np.linspace(-np.pi + (np.pi * 0.75), np.pi - (np.pi * 0.25), 200)
y_arg = np.linspace(-np.pi + (np.pi * 0.75), np.pi - (np.pi * 0.25), 200)
plt.fill_between(
    x_arg, 2 * np.sin(y_arg + (np.pi * 0.25)), alpha=0.3, color="blue"
)

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols("x")
f = sp.sin(x) * sp.cos(x)

integral = sp.integrate(f, (x, 0, 2 * sp.pi))

print(f"Hasil integral: {integral}")
print(f"Hasil desimal: {float(integral)}")

x_interval = np.linspace(0, 2 * np.pi, 200)
y_vals = np.sin(x_interval) * np.cos(x_interval)

plt.figure(figsize=(8, 6))
plt.plot(x_interval, y_vals, color="blue")

x_first_arg = np.linspace(0, np.pi * 0.5, 200)
y_first_arg = np.linspace(0, np.pi * 0.5, 200)
plt.fill_between(x_first_arg, np.sin(y_first_arg) * np.cos(y_first_arg), alpha=0.3, color="blue")

x_second_arg = np.linspace(np.pi, (2 * np.pi) - (np.pi * 0.5) , 200)
y_second_arg = np.linspace(np.pi, (2 * np.pi) - (np.pi * 0.5) , 200)
plt.fill_between(x_second_arg, np.sin(y_first_arg) * np.cos(y_first_arg), alpha=0.3, color="blue")

plt.axhline(0)

plt.title("Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
